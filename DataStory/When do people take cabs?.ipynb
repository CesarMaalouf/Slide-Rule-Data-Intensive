## Cars are for parties

If we look at the days of the week that people hire cabs - both yellow cabs and uber cabs - we find something a little bit unexpected.  While rush hour traffic on weekdays might lead one to believe that weekdays would be the most common time to get a cab.  However, we see below that the more cabs are hailed on Friday and Saturday - the days with weekend nights - then the other days of the week.

While there is plenty of weekday usage of cabs, the spike on Friday and Saturday indicates that people like to hail cabs to go out on weekends the most.
